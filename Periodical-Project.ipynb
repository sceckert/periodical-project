```python
# Calculate the average percent composition of a given section heading
mean_percentage_of_topic_in_section_heading = df.pivot_table(index = 'section-heading', aggfunc='mean')
mean_percentage_of_topic_in_section_heading[:]
```

```python
nt_mean_percentage_of_topic_in_section_heading = mean_percentage_of_topic_in_section_heading.drop('year', axis=1)
nt_mean_percentage_of_topic_in_section_heading
```

```python
fixed_mean_percentage_of_topic_in_section_heading = nt_mean_percentage_of_topic_in_section_heading.drop('docId', axis=1)
fixed_mean_percentage_of_topic_in_section_heading
```

```python
# Select only the rows that have a certain section heading
section_heading_name_list = ['Fiction-and-Essays', 'Fiction', 'Poetry', 'Poetry-Fiction-Essays', 'Poetry-and-Fiction', 'Critical-Notices', 'General-Literature']
#Grab DataFrame rows where column has certain values
just_fiction = fixed_mean_percentage_of_topic_in_section_heading[fixed_mean_percentage_of_topic_in_section_heading.index.isin(section_heading_name_list)]
just_fiction
```

```python
over_one_percent = just_fiction[:] > 0.001
over_one_percent
```

```python
plt.figure()
```

```python
# Plot the section headings by topic composition
just_fiction.plot(kind='barh', stacked=True, colormap='Set3')
# Set x-axis limits
plt.xlim(0, 1.0)
# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
topics = fixed_mean_percentage_of_topic_in_section_heading[:0]
topics
```

```python
section_headings = fixed_mean_percentage_of_topic_in_section_heading.index
section_headings
```

```python
transposed_mean_percentage_of_topic_in_section_heading = mean_percentage_of_topic_in_section_heading.T[:]
transposed_mean_percentage_of_topic_in_section_heading
```

```python
transposed_mean_percent_of_topic_in_section_heading = transposed_mean_percentage_of_topic_in_section_heading.drop(['docId', 'year'])
transposed_mean_percent_of_topic_in_section_heading 
```

```python
# Configure the plot function
plt.figure(figsize=(10, 10))
```

```python
plt.pcolor(fixed_mean_percentage_of_topic_in_section_heading, cmap='Blues')
plt.yticks(np.arange(fixed_mean_percentage_of_topic_in_section_heading.shape[0])+0.5, section_headings);
plt.xticks(np.arange(fixed_mean_percentage_of_topic_in_section_heading.shape[1])+0.5, topics);
plt.xticks(rotation=90)
plt.ylabel('Section Headings')
plt.xlabel('Topics')
plt.title('Percentage of Topics in Section Heading')
plt.colorbar(cmap='Blues')
```

```python
plt.pcolor(transposed_mean_percent_of_topic_in_section_heading, cmap='Blues')
plt.yticks(np.arange(transposed_mean_percent_of_topic_in_section_heading.shape[0])+0.5, topics);
plt.xticks(np.arange(transposed_mean_percent_of_topic_in_section_heading.shape[1])+0.5, section_headings);
plt.xticks(rotation=90)
plt.ylabel('Topics')
plt.xlabel('Section Headings')
plt.title('Percentage of Topics by Section Heading (Tranposed)')
plt.colorbar(cmap='Blues')
```
