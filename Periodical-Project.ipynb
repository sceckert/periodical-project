```python
# Group a given topic: access a given topic label (eg "44 sea kind poetic")
# and call groupby with the value at a given "section-heading":
grouped = df['36 form sense power'].groupby(df['section-heading'])
```

```python
# Take the mean value of the group
grouped.mean()
```

```python
# Create a new data object that is the grouping of the percent composition of that topic
grouped_mean = grouped.mean()
```

```python
plt.figure()
```

```python
# Plot the mean values for a topic by section heading
grouped_mean.plot(kind='barh', rot=0)
plt.title('Share of Topic #36 form sense power')
```

```python
# Group a given topic: access a given topic label (eg "44 sea kind poetic")
# and call groupby with the value at a given "section-heading":
grouped = df['30 women character power'].groupby(df['section-heading'])
# Create a new data object that is the grouping of the percent composition of that topic
grouped_mean = grouped.mean()
plt.figure()
# Plot the mean values for a topic by section heading
grouped_mean.plot(kind='barh', stacked=True)
plt.title('Share of Topic #30 women character power')
```

```python
# Group a given topic: access a given topic label (eg "44 sea kind poetic")
# and call groupby with the value at a given "section-heading":
grouped = df['29 art picture artist'].groupby(df['year'])
# Create a new data object that is the grouping of the percent composition of that topic
grouped_mean = grouped.mean()
plt.figure(figsize=(10, 4))
# Plot the mean values for a topic by section heading
grouped_mean.plot(kind='bar', rot=0)
plt.title('Share of Topic #29 art picture artist')
```
